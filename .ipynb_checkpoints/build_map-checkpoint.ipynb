```python
import folium
from folium import Element

scotland_center = [56.4907, -4.2026]
m = folium.Map(location=scotland_center, zoom_start=6)

# 1. Add global styles to <head>
table_css = """
<style>
  table {
    border-collapse: collapse;
    font-family: Arial, sans-serif;
    box-shadow: 0 2px 8px rgba(0,0,0,0.1);
    background: #fff;
    table-layout: fixed;
  }
  th {
    border: 1px solid #ddd;
    background-color: #f7f7f7;
    text-align: left;
    font-weight: 600;
    height: auto;
  }
  td {
    border: 1px solid #ddd;
    text-align: left;
    height: auto;
  }
</style>
"""
m.get_root().header.add_child(Element(table_css))

# 2. Add JS filter script to <head>
table_script = """
<script>
  document.addEventListener("DOMContentLoaded", function () {
    const searchInput = document.getElementById('searchInput');
    const rows = document.querySelectorAll('#mountainTable tbody tr');
    searchInput.addEventListener('input', function () {
      const filter = searchInput.value.toLowerCase();
      rows.forEach(row => {
        const nameText = row.cells[0].textContent.toLowerCase();
        row.style.display = nameText.includes(filter) ? '' : 'none';
      });
    });
  });
</script>
"""

m.get_root().header.add_child(Element(table_script))


sidebar_html = f"""
            <div id="sidebar" style="
              position: fixed;
              top: 0;
              left: 0;
              width: 170px;
              height: 100vh;
              background: #ffffff;
              overflow-y: auto;
              padding: 20px 15px;
              box-shadow: 2px 0 10px rgba(0, 0, 0, 0.1);
              z-index: 1000;
              font-family: 'Segoe UI', Tahoma, sans-serif;
              font-size: 14px;
              line-height: 1.6;
              color: #333;
              border-right: 1px solid #ddd;
            ">
              <table id="mountainTable">
                <colgroup><col style="width: 100px"></colgroup>
                <thead>
                  <tr>
                    <th>
                      Name<br>
                      <input type="text" id="searchInput" placeholder="🔍" style="width: 100px">
                    </th>
                  </tr>
                </thead>
                <tbody>
                  {table_rows}
                </tbody>
              </table>
            </div>
            """

m.get_root().html.add_child(Element(sidebar_html))

# 4. Shift the folium map to the right of the sidebar
map_div_id = m.get_name()
map_css = f"""
<style>
  #{map_div_id} {{
    position: absolute;
    top: 0;
    left: 180;
    right: 0;
    bottom: 0;
    z-index: 0;
  }}
</style>
"""
m.get_root().header.add_child(Element(map_css))

assign_map_js = f"""
<script>
  window.onload = function() {{
    window.map = {map_div_id};
    window.panMap = function(lat, lng) {{
      window.map.setView([lat, lng], 10);
    }};
  }};
</script>
"""
m.get_root().html.add_child(Element(assign_map_js))






# Save
m.save("map_with_sidebar_table.html")
```

```python
import folium
from folium import Element

# Create Folium map
scotland_center = [56.4907, -4.2026]
m = folium.Map(location=scotland_center, zoom_start=6)

# Sidebar HTML (absolute positioned)
sidebar_html_template = Template("""
<div id="sidebar" style="
  position: fixed;
  top: 0;
  left: 0;
  width: 300px;
  height: 100vh;
  background: #f4f4f4;
  overflow-y: auto;
  padding: 15px;
  box-shadow: 2px 0 5px rgba(0,0,0,0.1);
  z-index: 1000;
">
  $table
</div>
""")

sidebar_html = sidebar_html_template.substitute({"table": table_html})

# Add sidebar to HTML
m.get_root().html.add_child(Element(sidebar_html))

# Get map container ID (e.g. "map_ab123...")
map_div_id = m.get_name()

# Shift map to the right of sidebar using its actual ID
map_css = f"""
<style>
  #{map_div_id} {{
    position: absolute;
    top: 0;
    left: 300px;  /* match sidebar width */
    right: 0;
    bottom: 0;
    z-index: 0;
  }}
</style>
"""
m.get_root().header.add_child(Element(map_css))

# JS: Wait until window load to assign map + define panMap
map_js = f"""
<script>
  window.onload = function() {{
    window.map = {map_div_id};
    window.panMap = function(lat, lng) {{
      window.map.setView([lat, lng], 10);
    }};
  }};
</script>
"""
m.get_root().html.add_child(Element(map_js))

# Save map to HTML file
m.save("map_with_sidebar_fixed.html")
```

```python
def add_marker(row):

    name = row["Name"]
    height = row["Metres"]
    coords = [row["Latitude"], row["Longitude"]]
    year = row["Matthew"]

    if year != "":
        icon = bag_icon
        label = "<br>".join([name, f"height: {height}m", f"bagged: {year}"])
    else:
        icon = no_bag_icon
        label = "<br>".join([name, f"height: {height}m"])
        
    folium.Marker(
    location = coords,
    tooltip = label,
    icon = icon
    ).add_to(m)

    return
```
